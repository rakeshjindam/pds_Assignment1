```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

data= pd.read_csv('/content/StudentsPerformance.csv')
display(data.head(10))
```

```python

df=data['race/ethnicity'] # creating seperate data frame for the race/ethnicity attribute
df.value_counts().plot(kind='pie') # using value_counts() method to count various groups and plot a pie chart
plt.axis('equal')
plt.title('Pie chart of various groups')
```

```python

data.gender.value_counts().plot(kind='barh') #Using value.count() method and plotting a bar graph
plt.ylabel('Gender')
plt.xlabel('Count')
plt.title('Gender count')
```

```python

df1=pd.DataFrame(data,columns=["gender","math score","reading score","writing score"]) #creating a seperate data frame with the given attributes
malemask=df1["gender"]=="male"  #creating a mask to seperate males and females
dfmale = df1[malemask]

dfmaleavg=dfmale[["math score","reading score","writing score"]].mean() # .mean() method is used for mean of all values

dffemale = df1[~malemask]


dffemaleavg=dffemale[["math score","reading score","writing score"]].mean()

r1=[1,3,5] #r1 and r2 are used for position of bar plot
r2=[2,4,6]
plt.bar(r1,dfmaleavg) 
plt.bar(r2,dffemaleavg) # using r1 and r2 a combined bar plot visualization is done
xt=("math","reading","writing") #for labeling each set of plots

plt.xticks([1.5,3.5,5.5],xt) # for position of plots
plt.legend(["male","female"],bbox_to_anchor=(1.25,0.9),loc="center right") # to position the legend
plt.xlabel("Areas of score",fontsize="15") # x and y axis labeling
plt.ylabel("Score",fontsize="15")
plt.title("Comparision of male and female scores in subjects",fontsize="20")
plt.show()
```

```python
plt.figure(figsize=(10,10)) # using figsize() method for the size of figure
plt.hist(data['parental level of education'],edgecolor='black',bins=6) # to plot a histogram
xt1=("bachelor's degree","some collage","masters degree","associates degree","high school","some high school") # labels for histogram
plt.xticks([0.35,1.25,2.05,2.95,3.8,4.6],xt1) # position of each plot in histogram
plt.xlabel('Various Degrees',fontsize="15")
plt.ylabel('Count',fontsize="15")
plt.title('Histogram For Parental Level Of Education',fontsize="15")
plt.show()
```

```python

plt.figure(figsize=(10,10)) #for the size of visualization
df2=pd.DataFrame(data, columns=['parental level of education','math score','reading score','writing score']) # creating a dataframe
df2['mean']=df2.loc[:,["math score","reading score","writing score"]].mean(axis=1) # creating new attribute mean which is avg of 3 scores
plt.scatter(df2['parental level of education'],df2['mean']) # scatter plot against mean and parental education
xt1=("bachelor's degree","some collage","masters degree","associates degree","high school","some high school") # plot labels
plt.xticks([0.,1,2,3,4,5],xt1) # plot positions
plt.xlabel(" Parental Education",fontsize="14")
plt.ylabel(" mean of all scoring areas",fontsize="14")
plt.title(" Scatter plot for student score based on parental level of education ",fontsize="18")
plt.show()
```
